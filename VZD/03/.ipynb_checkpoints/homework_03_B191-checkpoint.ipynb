# Úkol č. 3 - Segmentace zákazníků e-shopu (do 29. listopadu)

Jednou z důležitých aplikací shlukování je **segmentace zákazníků** (angl. **customer segmentation**). 

Předpokládejme, že máme následující obchodní údaje o prodejích (resp. nákupech z pohledu zákazníků):
TransactionID - ID nákupu,
CustomerID - ID zákazníka, 
Date - datum nákupu, 
Total - celková cena nákupu.

Chceme najít segmenty zákazníků, kteří se chovají podobně. K tomu je dobré informace z jednotlivých nákupů pro individuální zákazníky agregovat. Tj. získat pro každého zákazníka jeden řádek.

Populárním přístupem je **RFM**, což znamená:

- **R**ecency: Počet dnů od posledního nákupu (poslední datum v datasetu pro daného zákazníka).
    - Počet dnů počítejte ke dni uskutečnění poslední transakce v celém datasetu (tj. 12/19/2015), nikoli k dnešku. Tváříme se, že jde o aktuální data.
- **F**requency: Počet nákupů. Občas se vynechávají zákazníci s jediným nákupem. Pro jednoduchost je zde ale necháme.
- **M**onetary: Celková suma, kterou daný zákazník utratil.

**Vytvoření RFM data framu**

```python
df2 = df.copy()

# RECENCY
# Set a fixed today date as advised in the task description
today = pd.datetime(2015, 12, 19)
# to determine Recency we will need just the Date column
recency = df2.drop(['Subtotal', 'Country'], axis=1)
# Convert date values to number of days which passed since the date until today
recency['Date'] = (today - pd.to_datetime(recency['Date'])).dt.days
#Take the minimum of days (the most recent order)
recency = recency.groupby(['Customer ID'])['Date'].min();

# FREQUENCY
# when I group by Customer ID and count number of elements in each group I obtain frequency for each customer
frequency  = df2.groupby(['Customer ID'])['Date'].count()

# MONETARY
# monetary column is sum of all subtotals for given customer
monetary = df2.groupby(['Customer ID'])['Subtotal'].sum()

# display(recency.head())
# display(frequency.head())
# display(monetary.head())

# create RFM dataframe
rfm_df = pd.concat([recency, frequency, monetary], axis=1)
rfm_df.columns = ['Recency' , 'Frequency', 'Monetary']
rfm_df.head()
```

**Normalizace příznaků**
* Škálování určitě ovlivní výsledky k-means algoritmu ve velké míře a je vhodné ho pro tento problem využít
* Jelikož jsou jednotlivé sloupce mezi sebou neporovnatelné, rozhodl jsem se u nich provést normalizaci
    * hlavně _Recency_ sloupec způsoboval, že se vytvářené shluky shromažďovaly do jedné přímky, neboli "roztahoval shluky" a jako následek toho ve výsledných shlucích pak nebyl patrný projev ostatních parametrů (frequency a monetary) a nebylo možné z grafu skoro nic vyčíst
* Rozhodl jsem se tedy použít _sklearn.preprocessing.StandardScaler_ . Po jeho použití bylo v jednotlivych shlukách vidět i projevy ostatních parametrů než jen _Recency_

```python
from sklearn.preprocessing import StandardScaler

standart_scaler = StandardScaler().fit(rfm_df)
rfm_norm = standart_scaler.transform(rfm_df)
display(rfm_norm)
```

Pro určení neoptimálnějšího _k_ pro **k-means** algoritmus použiji **elbow method**
* Nejdříve vyhledám nejlepší parametr _k_ pro k-means algoritmus

```python
from sklearn.cluster import KMeans

ix = np.zeros(21)
iy = np.zeros(21)
# repeat alhgorithm k-means and save the sums of squared distances to closest cluster centers to iy
# for usage in the elbow method
for k in range(ix.shape[0]):
    kmeans = KMeans(n_clusters=k+1, random_state = 666)
    kmeans.fit(rfm_norm)
    iy[k] = kmeans.inertia_
    ix[k] = k+1
```

Vizualizace elbow methody

```python
plt.figure(figsize=(20, 10))
plt.xlabel('parameter k')
plt.ylabel('Objective function')
plt.plot(ix, iy, 'o-')
plt.show()
```

* Ve vizualiazi můžeme vidět prudkou změnu směru funkce, kolem bodu 3 a 4
* V bodě 3 mi přijde změna směru funkce nejprudčí, proto jse se nakonec rozhodl zvolit 3 shluky
    * zkoušel jsem použít i 4 shluky, ale přislo mi, že 2 ze shluků se navzájem velmi podobali, tak jsem se nakonec rozhodl, že mi stačí použít 3 shluky a popsat ty

```python
# clustering
k = 3
kmeans = KMeans(n_clusters = k, random_state = 666).fit(rfm_norm)
clusters = kmeans.labels_
print(clusters)

cluster_counts = [0, 0, 0]
for i in clusters:
    cluster_counts[i] += 1
print('\nPočty zákazníků v jednotlivých shlucích:')
for i in range(len(cluster_counts)):
    print(f'{i}: {cluster_counts[i]}')
```

**Vizualizace**
* Clustery jsem měl problém správně vizualizovat, aby byly správně vidět, tak jsem nechal graf vykreslit 2x z 2 různých pohledů

```python
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
fig1 = plt.figure(figsize=(10,10),facecolor='w', edgecolor='k')
fig2 = plt.figure(figsize=(10,10),facecolor='w', edgecolor='k')
ax1 = fig1.add_subplot(111, projection='3d')
ax2 = fig2.add_subplot(111, projection='3d')

for idx, row in enumerate(rfm_norm):
    recency   = row[0]
    frequency = row[1]
    monetary  = row[2]
    color = ['r','g','b'][clusters[idx]]
    mark = ['x','+','o','*'][clusters[idx]]
    ax1.scatter(recency, frequency, monetary, c=color, marker=mark)
    ax2.scatter(recency, frequency, monetary, c=color, marker=mark)

ax2.view_init(elev=10., azim=30)
    
ax1.set_xlabel('Recency')
ax1.set_ylabel('Frequency')
ax1.set_zlabel('Monetary')

ax2.set_xlabel('Recency')
ax2.set_ylabel('Frequency')
ax2.set_zlabel('Monetary')

plt.show()
```

**Popis:**
* **Zelené shluky** -> zákazníci, kteří jsou pro obchod dle mého názoru nejdůležitější
    * zákazníci, kteří celkově zaplatili největší sumy, mezi nimi mohou byt jak opakovaní zákazníci, tak jednorázoví
    * většina z nich jsou nedávní zákazníci
    * zákazniky v tomto shluku bych označil za **superstar zákazníky**

* **Červeno-modré shluky** -> méně zajímaví zakazníci
    * neopakovaní návštěvníci, kteří utratili minimální množství peněz
    * zákazníci, kteří párkrát obchod navštívili, ale vždy utratili jen malé množství peněz
    * rozdíl mezi modrými a červenými shluky je pouze v aktuálnosti jejich poslední návštěvy
    * jsou to zákazníci, kteří v obchodě utratili průměrné/střední množství peněz, ale jsou to zároveň vracející se a častí zákazníci
    * u červených shluků je vidět, že celkové utracené částky jsou lehce vyšší než u modrých shluků

## Nyní provedu stejný postup akorát pro modifikované RFM

Vytvoření modifikovaného RFM
* postupuji obdobně jako u vytváření původního RFM, rozdíly jsou okomentovány v kódu

```python
import math

df3 = df.copy()
df3['Date'] = (today - pd.to_datetime(df3['Date'])).dt.days

# RECENCY
recency_modif = df3.drop(['Subtotal', 'Country'], axis=1)
recency_modif = recency_modif.groupby(['Customer ID'])['Date'].min()
# number of months since last purchase is approximately
recency_modif = recency_modif.apply(lambda x: max(1,math.ceil(x/30.4167))) # 30.4167 -> average number of days in month

# FREQUENCY
# I need to groupBy on filtered dataset with -> date within one year
frequency_modif = df3[df3['Date'] <= 365].groupby(['Customer ID'])['Date'].count()

# MONETARY
# use max instead of sum
monetary_modif = df3.groupby(['Customer ID'])['Subtotal'].max()
monetary_modif = monetary_modif.apply(lambda x: max(1,x))

# create modified RFM dataframe
rfm_modif = pd.concat([recency_modif,frequency_modif,monetary_modif], axis=1)
rfm_modif.columns = ['Recency','Frequency','Monetary']
rfm_modif = rfm_modif.fillna(1)
display(rfm_modif.head())
```

Normalizace pomocí _StandardScaleru_

```python
standard_scaler_modif = StandardScaler().fit(rfm_modif)
rfm_norm_modif = standard_scaler_modif.transform(rfm_modif)
display(rfm_norm_modif)
```

Hledání nejlepšího **k** pro modifikovaný RFM a následná vizualizace _elbow methody_ 

```python
ix_modif = np.zeros(21)
iy_modif = np.zeros(21)
# repeat alhgorithm k-means and save the sums of squared distances to closest cluster centers to iy
# for usage in the elbow method
for k in range(ix_modif.shape[0]):
    kmeans = KMeans(n_clusters=k+1, random_state = 666)
    kmeans.fit(rfm_norm_modif)
    iy_modif[k] = kmeans.inertia_
    ix_modif[k] = k+1
```

```python
plt.figure(figsize=(20, 10));
plt.xlabel('parameter k')
plt.ylabel('Objective function')
plt.plot(ix_modif, iy_modif, 'o-')
plt.show()
```

Na základě této vizualizace jsem se rozhodl použít **k=4**

```python
# clustering
k = 4
kmeans = KMeans(n_clusters = k, random_state = 666).fit(rfm_norm_modif)
clusters_modif = kmeans.labels_
print(clusters_modif)

cluster_counts = [0, 0, 0, 0]
for i in clusters_modif:
    cluster_counts[i] += 1
print('\nPočty zákazníků v jednotlivých shlucích:')
for i in range(len(cluster_counts)):
    print(f'{i}: {cluster_counts[i]}')
```

**Výsledná vizualizace** pro modifikovaný RFM

```python
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(666)
fig1 = plt.figure(figsize=(10,10),facecolor='w', edgecolor='k')
fig2 = plt.figure(figsize=(10,10),facecolor='w', edgecolor='k')
ax1 = fig1.add_subplot(111, projection='3d')
ax2 = fig2.add_subplot(111, projection='3d')

for idx, row in enumerate(rfm_norm_modif):
    recency   = row[0]
    frequency = row[1]
    monetary  = row[2]
    color = ['r','g','b', 'y'][clusters_modif[idx]]
    mark = ['x','+','o','*'][clusters_modif[idx]]
    ax1.scatter(recency, frequency, monetary, c=color, marker=mark)
    ax2.scatter(recency, frequency, monetary, c=color, marker=mark)

ax2.view_init(elev=10., azim=30)
    
ax1.set_xlabel('Recency')
ax1.set_ylabel('Frequency')
ax1.set_zlabel('Monetary')

ax2.set_xlabel('Recency')
ax2.set_ylabel('Frequency')
ax2.set_zlabel('Monetary')

plt.show()
```

Při pohledu na vizualizaci jsem došel k těmto závěrům:

**Zelený shluk** -> nejzajímavější shluk/skupina zákazníků - super zákazníci
* vyskytuji se zde zákazníci s velkými útratami ať už pro jednorázovou návštěvu, nebo opakovanou

**Červené a modré shluky** -> méně zajímaví zákazníci
* **Červené shluky** -> tyto zákazníky bych zařadil do stálejších, běžnějších zákazníku, kteří sice nenakupují za velké částky, ale do obchodu se pravidelně vrací
    
* **Modré shluky** -> tito zákazníci obchod navštívili jednou/párkrát a utratili malé množství peněz -> celkově nejméně atraktivní shluk z pohlade majitele obchodu

**Žlutý shluk**
* ačkoliv mi podle elbow metody přišlo, že největší zlom je po 4 shlucích, tak mi do tohoto 4.tého shluku spadl jen 1 zákazník
* od ostatních se odlišujě tím, že do tohoto shluku by patřili opakovaní zákazníci, kteří se do obchodu vrací s velkými zakázkami. Tuto skupinu bych přirozeně rád označil za super zákazníky, ale vzhledem k tomu, že tam spadá jen 1 zákazník, tak to moc nemá smysl dle mého názoru
