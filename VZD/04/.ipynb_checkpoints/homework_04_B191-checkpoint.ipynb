```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import AdaBoostRegressor

import warnings
warnings.filterwarnings('ignore')
```

## Nejprve načtu data z excelove tabulky

```python
df = pd.read_excel("Residential-Building-Data-Set.xlsx", header=[0, 1], sheet_name='Data')
df.head()
```

```python
# df.iloc[:, df.columns.get_level_values(1)=='V-20']
```

### Kontrola zda neexistuji nejake nevyplnene hodnoty v nekterych radcich

```python
print(f'Are some values missing? {len(df.loc[:,df.isnull().sum() > 0].isnull().sum()) != 0}')
```

### Po pohledu na data jsem se rozhodl smazat vsechny sloupce V-20 a 23, protoze obsahuju velke mnozstvi podobnych hodnot
* V tomto kroku take dropnu prvni radek obsahujici nadpisy
* Take prejmenuji vysvetlovane hodnoty aby lepe vysvetlovali hodnotu, kterou obsahuji

```python
for x in ['20', '23']:
    df.drop(columns=df.iloc[:, df.columns.get_level_values(1)==('V-' + x)], inplace=True)

df = df.droplevel(0, axis=1)

df.rename({'V-9': 'Sales prices'}, inplace=True, axis=1)
df.rename({'V-10': 'Construction cost'}, inplace=True, axis=1)
# df.head()
```

### Rozdeleni dat na testovaci a trenovaci mnozinu
* Data jsem rozdelil na 3 casti (samotna data, hodnoty pro `sales prices` a hodnoty pro `construction costs`) a z techto casti jsem vytvoril sadu testovacich a trenovacich dat

```python
X = df.drop(columns=['Sales prices', 'Construction cost'])
sales_prices = df['Sales prices']
construction_cost = df['Construction cost']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_sales_prices_train, Y_sales_prices_test, Y_construction_cost_train, Y_construction_cost_test = train_test_split(X, sales_prices, construction_cost, test_size=0.25, random_state=21)
```

### Pro vetsi prehlednost kodu jsem nakonec vytvoril zvlast funkce pro vykonavani linearni a hrebenove regrese, funkci pro ziskavani nejlepsich parametru pro libovolny model a funkci pro plotovani vysledku

```python
def get_best_params(model, params, X, Y):
    
    GSCV = GridSearchCV(estimator=model, param_grid=params, scoring='neg_mean_absolute_error', iid=False, verbose=0)
    GSCV.fit(X, Y)
#     print('Best params:', GSCV.best_params_)
#     print('Score:', GSCV.best_score_)
    return GSCV.best_params_
```

```python
def linear_regression(X_train, X_test, Y_train, Y_test):
    params = {'fit_intercept': [True, False],
              'normalize': [True, False]}
    
    best_params = get_best_params(LinearRegression(), params, X_train, Y_train)
    print(f'Best params for linear regression: {best_params}')

    lr = LinearRegression(**best_params)
    lr.fit(X_train, Y_train)
    Y_pred = lr.predict(X_test)

    mean_absolute_error_value = mean_absolute_error(Y_test, Y_pred)
    print('Linear regression mean absolute error: ', mean_absolute_error(Y_test, Y_pred))
    
    plot_results(Y_test, Y_pred, 'Linear regression')
```

```python
# funkce pro plotovani vysledku
def plot_results(Y_test, Y_pred, label):
    plt.scatter(Y_test, Y_pred)
    plt.plot([0,max(Y_test)],[0,max(Y_test)], color='black')
    plt.title(label)
    plt.xlabel('Y_test')
    plt.ylabel('Y_pred')
    plt.show()
```

```python
def ridge_regression(X_train, X_test, Y_train, Y_test):
    params = {'alpha': [0.00000001, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 1.21, 2.1, 3.5, 4, 5, 6.6, 10, 21, 44, 66, 1000],
              'fit_intercept': [True, False],
              'normalize': [True, False]}
    
    best_params = get_best_params(Ridge(), params, X_train, Y_train)
    print(f'Best params for ridge regression: {best_params}')
    
    ridge = Ridge(**best_params)
    ridge.fit(X_train, Y_train)
    Y_pred = ridge.predict(X_test)

    mean_absolute_error_value = mean_absolute_error(Y_test, Y_pred)
    print('Ridge regression mean absolute error: ', mean_absolute_error(Y_test, Y_pred))
    
    plot_results(Y_test, Y_pred, 'Ridge regression')
```

### Aplikovani linearni regrese na sales prices

```python
linear_regression(X_train, X_test, Y_sales_prices_train, Y_sales_prices_test)
```

### Aplikovani linearni regrese na construction cost

```python
linear_regression(X_train, X_test, Y_construction_cost_train, Y_construction_cost_test)
```

### Aplikovani hrebenove regrese na sales prices

```python
ridge_regression(X_train, X_test, Y_sales_prices_train, Y_sales_prices_test)
```

### Aplikovani hrebenove regrese na construction cost

```python
ridge_regression(X_train, X_test, Y_construction_cost_train, Y_construction_cost_test)
```

### Po zkousenich ruznych nastroju na normalizaci jsem nedochazel k zlepšení výsledků. Pouze při použití StandartScalleru jsem zaznamenal lehce lepší ci srovnatelne výsledky u hřebenové regresse.

```python
from sklearn.preprocessing import StandardScaler

standart_scaler = StandardScaler().fit(X_train)
normalized_train = standart_scaler.transform(X_train)
normalized_test = standart_scaler.transform(X_test)

# print('Linear regression:') # -> horsi vysledky
# print('Sales prices:')
# linear_regression(normalized_train, normalized_test, Y_sales_prices_train, Y_sales_prices_test)
# print('Construction cost:')
# linear_regression(normalized_train, normalized_test, Y_construction_cost_train, Y_construction_cost_test)

# print('-----------------------------------------------------------')
# print('-----------------------------------------------------------')

print('Ridge regression:')
print('Sales prices:')
ridge_regression(normalized_train, normalized_test, Y_sales_prices_train, Y_sales_prices_test)
print('Construction cost:')
ridge_regression(normalized_train, normalized_test, Y_construction_cost_train, Y_construction_cost_test)
```

### Jako jiny model predikce jsem se rozhodl pouzit AdaboostRegressor

```python
def ada_boost_regressor(X_train, X_test, Y_train, Y_test):
    ada = AdaBoostRegressor()
    ada.fit(X_train, Y_train)
    Y_pred = ada.predict(X_test)
    
    mean_absolute_error_value = mean_absolute_error(Y_test, Y_pred)
    print(f'AdaBoost mean absolute error: {mean_absolute_error_value}')
    
    plot_results(Y_test, Y_pred, 'AdaBoost')
    
    return mean_absolute_error_value
```

```python
print('Sales prices:')
ada_boost_regressor(X_train, X_test, Y_sales_prices_train, Y_sales_prices_test)
print('Construction cost:')
ada_boost_regressor(X_train, X_test, Y_construction_cost_train, Y_construction_cost_test)
```

### Vytvoreni novych priznaku
* Pri premysleni nad moznostmi jake nove smysluplne priznaky bych mohl vytvorit me napadly dva nove priznaky a to:
    * `Construction time`, ktery zachycuje pocet let od pocatku stavby az do konce
    * `Object cost`, ktery vznikne vynasobenim prostoru objektu s cenou za metr ctverecni objektu
* Aplikoval sem na rozsirena data stejne kroky jako jsem aplikoval na data puvodni v krocich vyse

```python
df_extended = df.copy()
df_extended['Construction time'] = df_extended['COMPLETION YEAR'] - df_extended['START YEAR'] + (df_extended['COMPLETION QUARTER'] - df_extended['START QUARTER'])/4
# floor area * price per m^2
df_extended['Object cost'] = df_extended['V-2'] * df_extended['V-8']
# df_extended.head()
```

```python
X_extended = df_extended.drop(columns=['Sales prices', 'Construction cost'])
sales_prices = df_extended['Sales prices']
construction_cost = df_extended['Construction cost']

X_extended_train, X_extended_test, Y_sales_prices_train, Y_sales_prices_test, Y_construction_cost_train, Y_construction_cost_test = train_test_split(X_extended, sales_prices, construction_cost, test_size=0.25, random_state=21)
```

### Linearni regrese

```python
print('Sales Prices')
linear_regression(X_extended_train, X_extended_test, Y_sales_prices_train, Y_sales_prices_test)
print('Construction Cost')
linear_regression(X_extended_train, X_extended_test, Y_construction_cost_train, Y_construction_cost_test)
```

### Hrebenova regrese

```python
print('Sales Prices')
ridge_regression(X_extended_train, X_extended_test, Y_sales_prices_train, Y_sales_prices_test)
print('Construction Cost')
ridge_regression(X_extended_train, X_extended_test, Y_construction_cost_train, Y_construction_cost_test)
```
